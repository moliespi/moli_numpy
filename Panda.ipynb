SERIES

```python
#Crear una serie
serie=pd.Series([2,6,-1,8,9,2.3], index=['a','b','c','d','e','f']) # la cantidad de numeros debe ser proporcional a la de letras
serie
```

```python
#Borrar series o filas
serie2=serie.drop(['a','e'])
serie2
```

```python
#Seleccion de filas o series
serie[['b','d']]
```

```python
#Asignación
serie['b']=100
serie
```

```python
#OPERADORES
serie4=serie+serie2
serie4
```

```python
serie5=serie-serie2
serie5
```

```python
serie6=serie*serie2
serie6
```
